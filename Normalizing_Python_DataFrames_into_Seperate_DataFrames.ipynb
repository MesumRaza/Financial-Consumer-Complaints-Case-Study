```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/MesumRaza/Financial-Consumer-Complaints-Case-Study/main/Consumer_Complaints_sampled.csv')
```

```python
df.head()
```

```python
state_factor=pd.factorize(df.State)
df['state_id']=state_factor[0]+1
df.drop(['State'], axis=1)
```

```python
df.head()
```

```python
state_df=pd.DataFrame({'state_id':range(1,state_factor[1].size+1),'state_name':state_factor[1]})
```

```python
state_df
```

```python
Product_factor=pd.factorize(df.Product)
df['Product_id']=Product_factor[0]+1
df.drop(['Product'], axis=1,inplace=True)
Product_df=pd.DataFrame({'Product_id':range(1,Product_factor[1].size+1),'Product_name':Product_factor[1]})
```
